## Understand the working albumentation using single image

There could be several reasons why the bounding box is not being displayed correctly in the augmented image. Here are some possible reasons:

Incorrect normalization: Make sure that you are using the correct normalization values when converting the bounding box coordinates from the YOLO format to pixel values. If the normalization factors are incorrect, the bounding box will be placed at the wrong location.

Incorrect order of coordinates: Make sure that you are using the correct order of coordinates when drawing the bounding box. In the YOLO format, the coordinates are usually given as (center_x, center_y, width, height), while in some other formats, the coordinates are given as (xmin, ymin, xmax, ymax). If you are using the wrong order of coordinates, the bounding box will be drawn at the wrong location.

Incorrect image size: Make sure that the size of the augmented image is the same as the size of the original image. If the size of the augmented image is different, the bounding box coordinates will need to be scaled accordingly.

Incorrect index of bounding box: If you are working with multiple bounding boxes, make sure that you are using the correct index of the bounding box when drawing the rectangle. If you are using the wrong index, the bounding box will be drawn at the wrong location.

Albumentations requires the bounding box annotations to be in the format of a list of dictionaries, where each dictionary contains the xmin, ymin, xmax, ymax, and label keys.

YOLO format annotations consist of a text file for each image, where each line in the file represents an object in the image and contains the label, x_center, y_center, width, and height values.

Therefore, to use Albumentations with YOLO format annotations, you need to first read in the YOLO format annotations and convert them to the format that Albumentations expects. After applying data augmentations, you will need to convert the augmented annotations back to YOLO format if you want to use them with YOLO-based models.

You can use the yolo_to_alb and alb_to_yolo functions in my previous answer to convert between the YOLO and Albumentations annotation formats.

```python
# Function to convert YOLO format annotations to albumentation format
def yolo_to_alb(annotations, image_width, image_height):
    
    # Empty list to store the annotations in albumentation format 
    alb_annot = []
    x, y, w, h = annotations[1:]
    xmin = int((x - w/2) * image_width)
    xmax = int((x + w/2) * image_width)
    ymin = int((y - h/2) * image_height)
    ymax = int((y + h/2) * image_height)
    label = int(annotations[0]) # Label of the annotated image
        
    # Let's append the annotations in the alb_annot list
    alb_annot.append({'xmin': xmin,'ymin': ymin,'xmax': xmax,'ymax': ymax,'label': label})
    
    return alb_annot
```

```python
# Define a function to convert Albumentations format annotations to YOLO format
def alb_to_yolo(annot, image_width, image_height):
    
    # Empty list to store the YOLO format coorinates
    yolo_annot = []

    xmin = annot[0]['xmin']
    ymin = annot[0]['ymin']
    xmax = annot[0]['xmax']
    ymax = annot[0]['ymax']
    width = xmax - xmin
    height = ymax - ymin
    x_center = (xmin + xmax) / 2 / image_width
    y_center = (ymin + ymax) / 2 / image_height
    label = annot[0]['label']
    yolo_annot.append([label, x_center, y_center, width / image_width, height / image_height])
    
    return yolo_annot
```

# Data augmentation pipeline

Now let's create a data augmnetation pipeline

```python
for folder in ['train','test','validation']: 
    for image in os.listdir(os.path.join('Drowsy_data', folder, 'images')):
        img = cv2.imread(os.path.join('Drowsy_data', folder, 'images', image)) # Reading an image
        
        # We have initialized a list having coordinate values in the case there is no face in the image
        coords = [0,0,0.00001,0.00001]
        
        # Reading label from the specified label path
        label_path = os.path.join('Drowsy_data', folder, 'labels', f'{image.split(".")[0]}.txt')
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                label = np.loadtxt(f)

            # Converting the annotations from YOLO format to albumentation format
            alb_format = yolo_to_alb(label,640,480)
            
            # Storing the coordinates from the label dictionary, corrdinates are : X_min,Y_min,X_max,Y_max
            coords[0] = alb_format[0]['xmin']
            coords[1] = alb_format[0]['ymin']
            coords[2] = alb_format[0]['xmax']
            coords[3] = alb_format[0]['ymax']
            coords = list(np.divide(coords, [640,480,640,480])) # Nomalizing the coordinate values

        try: 
            for x in range(60): # From a base image we will create 60 augmented images
                class_labels = [['Drowsy', 'Awake']]
                augmented = augmentor(image=img, bboxes=[coords], class_labels=class_labels)
                
                # We will store the augmented images in the below mentioned folder
                cv2.imwrite(os.path.join('Aug_data', folder, 'images', f'{image.split(".")[0]}.{x}.jpg'), augmented['image'])

                annotation = {}
                annotation['image'] = image

                if os.path.exists(label_path):
                    if len(augmented['bboxes']) == 0: 
                        annotation['bbox'] = [0,0,0,0]
                        annotation['class'] = 0 
                    else: 
                        annotation['bbox'] = augmented['bboxes'][0]
                        annotation['class'] = 1
                else: 
                    annotation['bbox'] = [0,0,0,0]
                    annotation['class'] = 0 


                with open(os.path.join('Aug_data', folder, 'labels', f'{image.split(".")[0]}.{x}.json'), 'w') as f:
                    json.dump(annotation, f)

        except Exception as e:
            print(e)
```

```python
# Importing layers and base network
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Dense, GlobalMaxPooling2D
from tensorflow.keras.applications import VGG16

# Downloading VGG16 model
vgg = VGG16(include_top=False)
```

```python
vgg.summary()
```

```python
# Let's now build the model
def build_model(): 
    input_layer = Input(shape=(120,120,3))
    
    vgg = VGG16(include_top=False)(input_layer)

    # Classification Model  
    f1 = GlobalMaxPooling2D()(vgg)
    class1 = Dense(2048, activation='relu')(f1)
    class2 = Dense(1, activation='sigmoid')(class1)
    
    # Bounding box model
    f2 = GlobalMaxPooling2D()(vgg)
    regress1 = Dense(2048, activation='relu')(f2)
    regress2 = Dense(4, activation='sigmoid')(regress1)
    
    facetracker = Model(inputs=input_layer, outputs=[class2, regress2])
    return facetracker
```

```python
facetracker = build_model()
facetracker.summary()
```

# Defining loss and optimizers

```python
batches_per_epoch = len(train)
lr_decay = (1./0.75 -1)/batches_per_epoch
opt = tf.keras.optimizers.Adam(learning_rate=0.0001, decay=lr_decay)
```

```python
def localization_loss(y_true, yhat):            
    delta_coord = tf.reduce_sum(tf.square(y_true[:,:2] - yhat[:,:2]))
                  
    h_true = y_true[:,3] - y_true[:,1] 
    w_true = y_true[:,2] - y_true[:,0] 

    h_pred = yhat[:,3] - yhat[:,1] 
    w_pred = yhat[:,2] - yhat[:,0] 
    
    delta_size = tf.reduce_sum(tf.square(w_true - w_pred) + tf.square(h_true-h_pred))
    
    return delta_coord + delta_size
```

```python
classloss = tf.keras.losses.BinaryCrossentropy()
regressloss = localization_loss
```

```python
class FaceTracker(Model): 
    def __init__(self, eyetracker,  **kwargs): 
        super().__init__(**kwargs)
        self.model = eyetracker

    def compile(self, opt, classloss, localizationloss, **kwargs):
        super().compile(**kwargs)
        self.closs = classloss
        self.lloss = localizationloss
        self.opt = opt
    
    def train_step(self, batch, **kwargs): 
        
        X, y = batch
        
        with tf.GradientTape() as tape: 
            classes, coords = self.model(X, training=True)
            
            batch_classloss = self.closs(y[0], classes)
            batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)
            
            total_loss = batch_localizationloss+0.5*batch_classloss
            
            grad = tape.gradient(total_loss, self.model.trainable_variables)
        
        opt.apply_gradients(zip(grad, self.model.trainable_variables))
        
        return {"total_loss":total_loss, "class_loss":batch_classloss, "regress_loss":batch_localizationloss}
    
    def test_step(self, batch, **kwargs): 
        X, y = batch
        
        classes, coords = self.model(X, training=False)
        
        batch_classloss = self.closs(y[0], classes)
        batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)
        total_loss = batch_localizationloss+0.5*batch_classloss
        
        return {"total_loss":total_loss, "class_loss":batch_classloss, "regress_loss":batch_localizationloss}
        
    def call(self, X, **kwargs): 
        return self.model(X, **kwargs)
```

```python
model = FaceTracker(facetracker)
model.compile(opt, classloss, regressloss)
```

# Training the model 

```python
logdir='logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
hist = model.fit(train, epochs=40, validation_data=val, callbacks=[tensorboard_callback])
```
